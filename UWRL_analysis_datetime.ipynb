```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# df = pd.read_csv('UWRL_river_velocimetry_dataset.csv')
df = pd.read_csv('UWRL_river_velocimetry_dataset_NEW.csv')
```

```python
df['v_pos'] =  [0 if pd.isna(x) else (-1 if x < 0 else 1) for x in df['v_x']]
df.datetime = pd.to_datetime(df.datetime)
print(df.columns)
df
```

```python
df2 = df.groupby(['datetime', 'v_pos']).aggregate({'v_pos': 'count'})
df2 = df2.rename(columns={'v_pos': 'v_pos_counts'}).reset_index()
df2 
```

```python
plt.figure(figsize=(20, 10))

sns.lineplot(
    data=df2[df2.v_pos == -1],
    x='datetime',
    y='v_pos_counts',
    label='upstream'
)

sns.lineplot(
    data=df2[df2.v_pos == 0],
    x='datetime',
    y='v_pos_counts',
    label='null'
)

sns.lineplot(
    data=df2[df2.v_pos == 1],
    x='datetime',
    y='v_pos_counts',
    label='downstream'
)

plt.legend(title='vector direction', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()  # Adjust the plot to make room for the legend
plt.show()
```
