```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# df = pd.read_csv('UWRL_river_velocimetry_dataset.csv')
df = pd.read_csv('UWRL_river_velocimetry_dataset_NEW.csv')
```

```python
df['v_pos'] =  [0 if pd.isna(x) else (-1 if x < 0 else 1) for x in df['v_x']]
df.datetime = pd.to_datetime(df.datetime)
print(df.columns)
df
```

```python
df_ss = df.groupby(['vegetation', 'spacial_location', 'v_pos']).size().reset_index(name='counts')

for x in df_ss.vegetation.unique():

    print(x)

    sns.barplot(
        data=df_ss[df_ss.vegetation == x],
        x='spacial_location',
        y='counts',
        hue='v_pos'
    )
    plt.legend(labels=['downstream', 'NaN', 'upstream'], bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
    plt.close()
```

```python
sns.histplot(
    data=df,
    x='LRO_discharge',
    bins=10
)
```

```python
df.vegetation =  [0 if x == 'none' else (1 if x == 'light' else 2) for x in df.vegetation]
```

```python
sns.lineplot(
    data=df,
    x='datetime',
    y='LRO_discharge'
)

plt.show()
plt.close()

sns.lineplot(
    data=df,
    x='datetime',
    y='vegetation'
)

plt.show()
plt.close()
```

```python
sns.scatterplot(
    data=df,
    x='vegetation',
    y='LRO_discharge'    
)
```

```python
df["LRO_discharge_ss"] = [0 if x <= 10 else (1 if x <= 20 else 2) for x in df.LRO_discharge]

df_ss = df.groupby(['LRO_discharge_ss', 'spacial_location', 'v_pos']).size().reset_index(name='counts')

for x in df_ss.LRO_discharge_ss.unique():

    print(x)

    sns.barplot(
        data=df_ss[df_ss.LRO_discharge_ss == x],
        x='spacial_location',
        y='counts',
        hue='v_pos'
    )
    plt.legend(labels=['downstream', 'NaN', 'upstream'], bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
    plt.close()
```
